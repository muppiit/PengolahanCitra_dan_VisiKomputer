jawab:<br>

jawab :<br>
Perbedaan utama antara tipe data int16 dan int32 dalam pembuatan black image (gambar hitam) terletak pada rentang nilai yang bisa diwakili:

2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!

````python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)


img_rb = img_rgb.copy()
img_rb[:, :, 1] = 0

img_gb = img_rgb.copy()
img_gb[:, :, 0] = 0

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rb)
plt.title("Red-Blue")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_gb)
plt.title("Green-Blue")
plt.axis('off')

plt.show()
````

3. Tampilkan image baris ke 20-115, kolom 25-120!

````python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Crop the image to the specified rows and columns
cropped_img = img_rgb[20:116, 25:121]

# Display the cropped image
plt.figure(figsize=(5, 5))
plt.imshow(cropped_img)
plt.title("pixel (baris ke 20-115, kolom ke 25-120)")
plt.axis('off')
plt.show()
````

4. Tampilkan image baris ke 5-30, semua kolom, channel Red saja!

````python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Extract the red channel and crop to the specified rows
red_channel = img_rgb[5:31, :, 0]

# Display the cropped red channel image
plt.figure(figsize=(10, 5))
plt.imshow(red_channel, cmap='Reds')
plt.title("potongan gambar (baris ke 5-30, Red Channel)")
plt.axis('off')
plt.show()
````

5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image.
disarankan menggunakan bilangan acak/random!

````python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

np.random.seed(0)

image_with_boxes = img_rgb.copy()

for _ in range(5):
    width = np.random.randint(20, 100)
    height = np.random.randint(20, 100)

    x_start = np.random.randint(0, image_with_boxes.shape[1] - width)
    y_start = np.random.randint(0, image_with_boxes.shape[0] - height)

    color = tuple(np.random.randint(0, 256, size=3).tolist())

    cv.rectangle(image_with_boxes, (x_start, y_start), (x_start + width, y_start + height), color, -1)

plt.figure(figsize=(8, 8))
plt.imshow(image_with_boxes)
plt.title("foto dengan kotak random")
plt.axis('off')
plt.show()
````

6. Tampilkan image dengan posisi terbalik!

````python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

flipped_horizontal = cv.flip(img_rgb, 1)

flipped_vertical = cv.flip(img_rgb, 0)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(flipped_horizontal)
plt.title("Horizontal")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(flipped_vertical)
plt.title("Vertical")
plt.axis('off')

plt.show()
````
