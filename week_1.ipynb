Step 1

````python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
````

step 2

````python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg","https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg","https://placekitten.com/800/571"]
for url in urls:
  image = io.imread(url)
  image = cv.resize(image,(0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print("\n")
````

step 3

````python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
````

step 4

````python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[tinggi//2,y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
````

````python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)


for x in range(tinggi):
    image_3[x, lebar//2] = [255, 255, 255]

for i in range(min(lebar, tinggi)):
    image_4[i, i] = [255, 255, 255]
    image_4[tinggi-1-i, i] = [255, 255, 255]



final_frame = cv.hconcat((image_2, image_3))
final_frame2 = cv.hconcat((image_2, image_4))

cv2_imshow(final_frame,)
cv2_imshow(final_frame2)
````

````python
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

garis_panjang = 100
y_tengah = tinggi // 2
start_x = max(0, lebar // 2 - garis_panjang // 2)
end_x = min(lebar, lebar // 2 + garis_panjang // 2)

for x in range(start_x, end_x):
    image_5[y_tengah, x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_5))
cv2_imshow(final_frame)
````

````python
image_6 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

kotak_panjang = 100  # Panjang sisi kotak
kotak_lebar = 100    # Lebar sisi kotak
x_start = max(0, lebar // 2 - kotak_panjang // 2)
x_end = min(lebar, lebar // 2 + kotak_panjang // 2)
y_start = max(0, tinggi // 2 - kotak_lebar // 2)
y_end = min(tinggi, tinggi // 2 + kotak_lebar // 2)

# Menggambar kotak pada image_2
image_6[y_start:y_end, x_start:x_end] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_6))
cv2_imshow(final_frame)
````
